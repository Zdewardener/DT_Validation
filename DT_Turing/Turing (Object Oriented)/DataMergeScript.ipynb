```python
import pandas as pd

combined_data = pd.DataFrame()  # Create an empty dataframe to store the combined data

for file in os.listdir('.'):  # Iterate over the files in the current directory
    if file.endswith('.csv'):  # Process only CSV files
        file_path = os.path.join('.', file)  # Get the full file path
        data = pd.read_csv(file_path)  # Read the CSV file
        combined_data = pd.concat([combined_data, data])  # Concatenate the data

# # Optionally, you can save the combined data to a new CSV file
# combined_data.to_csv('combined_data.csv', index=False)
```

```python
combined_data = combined_data.drop([ 'SourceID','SourceType', 'SourceLocation',
       'DestinationServiceAddress', 'DestinationServiceType',
       'DestinationServiceLocation', 'AccessedNodeType',
       'Operation'], axis=1)
```

```python
def move_column(dataframe, source_index, destination_index):
    columns = dataframe.columns.tolist()
    column_to_move = columns.pop(source_index)
    columns.insert(destination_index, column_to_move)
    return dataframe[columns]

source_index = 3  # Index of the column you want to move (0-based index)
destination_index = 1  # Index where you want to move the column (0-based index)

dataframe = move_column(combined_data, source_index, destination_index)

new_column_names = {'Time': 'Run Time',
                    'Value': 'msg',
                    'SourceAddress': 'node',
                    'AccessedNodeAddress': 'topics'}
df = dataframe.rename(columns=new_column_names)
df.to_csv('RWB_SimDataCombined.csv', index=False)
```
